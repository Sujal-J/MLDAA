```python
class Net:
    def __init__(self):        
        self.w1 = 0
        self.w2 = 0
        self.b = 0
        
    def forward(self, train):    
        wsum = self.w1 * train[0] + self.w2 * train[1] + self.b
        pred = 1 if wsum >= 0 else 0
        return (pred, wsum)

```
